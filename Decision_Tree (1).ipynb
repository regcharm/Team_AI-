```python


# 1. Load libraries, "sklearn" is the one used for machine learning
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report

# 2. Load data extracting information from the provided dataset
data = load_breast_cancer()
X = data.data
y = data.target

# 3. Data preprocessing using a ratio of 80% for training and 20% for testing
train_data, test_data, train_labels, test_label = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# 4. Model selection and training with the "fit" function
model = DecisionTreeClassifier(random_state=42)
model.fit(train_data, train_labels)

# 5. Model testing
predictions = model.predict(test_data)

# Muestra de resultados
print("--- Accuracy Score ---")
print(accuracy_score(test_label, predictions))
print("\n--- Classification Report ---")
print(classification_report(test_label, predictions, target_names=data.target_names))

# Visualización del árbol con un DPI = 300
plt.figure(figsize=(20,10), dpi=300)
plot_tree(model,
          feature_names=data.feature_names,
          class_names=data.target_names,
          filled=True,
          rounded=True)
plt.show()
```
